```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

df = pd.read_csv('cleanedData.csv').sample(frac=1)
df.head()
```

```python
ys = list(set(df.columns) - set(['ocean_proximityIndex']))
rows, cols = 2, 3
pointer = 0
figure, axes = plt.subplots(rows, cols, figsize=(15, 15))
# title for entire figure
figure.suptitle("Ocean proximity vs ...", fontsize=25, color="magenta")
for i in range(rows):
    for j in range(cols):
        sb.boxplot(ax=axes[i, j], data=df, x='ocean_proximityIndex', y=ys[pointer])
        axes[i, j].set_title(ys[pointer], fontsize=25, color="green")
        pointer += 1
plt.savefig('Ocean proximity index vs.png')
```

Comments:
* Ocean proximity has low correlation with the number of bedrooms as there is just slight discrepancy of bedrooms among proximities.
* Generally, this is not surprising since the closer to coast, the higher value a house is expected to have. The case of proximity 1.0 can be explained that if a house is not included in an area with a certain radius range, the effect of proximity fades.
* This is quite surprising since the median income is expected to be high in houses closer to coast. We wonder how median income was computed. A guess is either rich people are good at hiding their income.
* Similar to the number of bedrooms, the correlation between ocean proximity and population or households is quite low.
* It is understandable when the housing median age tends to be proportional with ocean priximity, which means that the elder people stay closer to ocean than younger ones due to the fact that senior citizens are richer than youngsters.
